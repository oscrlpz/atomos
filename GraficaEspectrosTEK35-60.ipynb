```julia
using PyPlot
```

```julia
archivosTodos=readdir("./");
archivosTek=filter(x->contains(x,"tek"), archivosTodos) #solo tomo archivos que incluyan la palabra "tek" en su nombre;
archivos=archivosTek[36:end] #me interesan los archivos 0008-0033
potencias=[(3.0:0.5:8.0);2.5;(2.5:-0.5:1.0);(0.8:-0.2:0.2);0.150;0.100;0.050;0.050;0.040;0.030];
```

```julia
N=length(potencias)
a=310;b=70; #elijo a y b para que la rampa de voltaje solo sea creciente
for n=1:N
    datos=readcsv("./"*archivos[n]);
    ch1=11.5+map(Float64,datos[1:end,2][14+a:end-1-b])
    ch2=11.5+map(Float64,datos[1:end,3][14+a:end-1-b])
    fluo=map(Float64,datos[1:end,4][14+a:end-1-b])
    volt=map(Float64,datos[1:end,5][14+a:end-1-b])
    
    figure()
    title((n,potencias[n]))
    plot(volt,(ch1-ch2),color="red")
    plot(volt,ch1,color="yellow")
    plot(volt,ch2,color="gray")
    plot(volt,fluo/10-0.2,color="purple")
    xlim([-0.25,0.25])
    ylim([-0.7,0.45])
end
```
